```python
# get user names
users = []
find_user = driver.find_elements(By.XPATH,'//*[@id="author-text"]/span')
for user_name in find_user:
        users.append(user_name.text)
users
```

```python
# time stamp
time_stamp = []
find_time=driver.find_elements(By.XPATH,'//*[@id="header-author"]/yt-formatted-string/a')
for time in find_time:
    time_stamp.append(time.text)
    
time_stamp
```

```python
likes = []

# get likes
find = driver.find_elements(By.XPATH,'//*[@id="vote-count-middle"]')
for like in find:
    likes.append(like.text)

likes
```

```python
#check
print(len(comments))
print(len(time_stamp))
print(len(users))
print(len(likes))
```

```python
# remove \n strings to avoid wrong display in output
comments_cleaned = []
for cmt in comments:
    clean_cmt = cmt.replace("\n","") #--> replace \n by whitespace
    clean_cmt.strip(" ") #--> remove unncessary whitespaces
    comments_cleaned.append(clean_cmt)
    
comments_cleaned[3]
```

```python
# store info in a dataframe
data = {'User': users, 'Comment': comments_cleaned, 'Time': time_stamp, 'Likes': likes}
df = pd.DataFrame(data)
df
```

```python
#sentence tokenization

from nltk import sent_tokenize

tokenized_sent = sent_tokenize(text)
print(tokenized_sent)
```

```python
#word tokenization
from nltk import word_tokenize

tokenized_word = word_tokenize(text)
print(tokenized_word)
```

```python
#remove .,'"
```

```python
stopwords = nltk.corpus.stopwords.words("english")
stopwords.append('.')
stopwords.append(',')
stopwords.append('!')
```

```python
#removal by stop word

final_words= []

for token in tokenized_word: 
    if token not in stopwords: 
        final_words.append(token) 

print(len(tokenized_word))
print(len(final_words))
```

```python
print(final_words)
```

```python
#!pip install scikit-learn
```

```python
#word_Frequency _
#bag of word
from sklearn.feature_extraction.text import TfidfVectorizer

tfidfv = TfidfVectorizer().fit(final_words)
tfidfv.vocabulary_
```

```python
tfidfv_dict = sorted(tfidfv.vocabulary_.items(),reverse =True,key= lambda x:x[1])
tfidfv_dict = dict(tfidfv_dict)
```

```python
tfidfv_df = pd.DataFrame.from_dict(tfidfv_dict,orient='index')
tfidfv_df
```

```python
#Check Positive and Negative by using SentimentintensityAnalyzer

dict_sent={}

for x in data['Comment']:
    sent_of_pos_neg = sia.polarity_scores(x)
    print(x,sent_of_pos_neg)
    if sent_of_pos_neg['neg']>0.1:
        dict_sent[x] = 'negative'
    elif sent_of_pos_neg['pos']>0.1:
        dict_sent[x] = 'positive'
    else:
        dict_sent[x] = 'neutral'
        
```

```python
dict_sent_afinn = {}

for x in data['Comment']:
    sent_of_pos_neg = afinn.score(x)
    dict_sent_afinn[x] = sent_of_pos_neg

sentiment_count = {}
for score in dict_sent_afinn.values():
    if score in sentiment_count:
        sentiment_count[score] += 1
    else:
        sentiment_count[score] = 1

scores = list(sentiment_count.keys())
counts = list(sentiment_count.values())

plt.bar(scores, counts)
plt.xlabel('Sentiment Score')
plt.ylabel('Count')
plt.title('Sentiment Analysis')
plt.show()
```
